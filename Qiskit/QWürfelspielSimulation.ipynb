```python
from qiskit.circuit.library import HGate
from qiskit import QuantumCircuit
from qiskit.transpiler import generate_preset_pass_manager
# Simulator
from qiskit_aer import AerSimulator

def wuerfeln():
    backend = AerSimulator(method='extended_stabilizer')
    hadamard_gate = HGate()

    # Circuit mit 3 Qubits erstellen
    circuit = QuantumCircuit(3, 3)
    for i in range(3):
        circuit.append(hadamard_gate, [i])
    circuit.measure([0, 1, 2], [0, 1, 2])

    pm = generate_preset_pass_manager(backend=backend)
    isa_circuit = pm.run(circuit)

    job = backend.run(isa_circuit, shots=1, memory=True)
    result = job.result()
    memory = result.get_memory()[0]  # z.â€¯B. '101'

    dezimal = int(memory, 2)

    if dezimal in [0, 7]:
        return wuerfeln()  # REKURSION
    else:
        return dezimal

def spiel_101_mit_risiko():
    punkte_spieler1 = 0
    punkte_spieler2 = 0
    aktueller_spieler = 1  # Spieler 1 beginnt

    while True:
        print(f"\nSpieler {aktueller_spieler} ist am Zug!")
        runden_punkte = 0
        zug_aktiv = True

        while zug_aktiv:
            eingabe = input("Zum WÃ¼rfeln 'Enter' drÃ¼cken oder '+' zum Sichern: ").strip().lower()
            
            if eingabe == '+':
                print(f"Spieler {aktueller_spieler} sichert {runden_punkte} Punkte.")
                if aktueller_spieler == 1:
                    punkte_spieler1 += runden_punkte
                else:
                    punkte_spieler2 += runden_punkte
                zug_aktiv = False
            else:
                wurf = wuerfeln()
                print(f"GewÃ¼rfelt: {wurf}")
                
                if wurf == 1:
                    if punkte_spieler1 == 100:
                        print("Spieler 1 hat gewonnen! ðŸŽ‰")
                        break
                    elif punkte_spieler2 == 100:
                        print("Spieler 2 hat gewonnen! ðŸŽ‰")
                        break
                    print(f"PECH! Spieler {aktueller_spieler} verliert alle Punkte dieser Runde! ðŸ˜±")
                    runden_punkte = 0
                    zug_aktiv = False
                elif (punkte_spieler1 + runden_punkte + wurf > 101) and aktueller_spieler == 1:
                    runden_punkte = 0
                    zug_aktiv = False
                    print("Mehr als 101 Punkte!")
                elif (punkte_spieler2 + runden_punkte + wurf > 101) and aktueller_spieler == 2:
                    runden_punkte = 0
                    zug_aktiv = False
                    print("Mehr als 101 Punkte!")
                else:
                    runden_punkte += wurf
                    print(f"Aktuelle Rundenpunkte: {runden_punkte} | Gesamt: {punkte_spieler1 + runden_punkte if aktueller_spieler == 1 else punkte_spieler2 + runden_punkte}")

        # Aktualisiere Gesamtpunkte (falls nicht schon durch '+' geschehen)
        if wurf != 1 and eingabe != '+':
            if aktueller_spieler == 1:
                punkte_spieler1 += runden_punkte
            else:
                punkte_spieler2 += runden_punkte

        print(f"Stand: Spieler 1 = {punkte_spieler1} | Spieler 2 = {punkte_spieler2}")

        # GewinnprÃ¼fung
        if punkte_spieler1 == 101:
            print("Spieler 1 hat gewonnen! ðŸŽ‰")
            break
        elif punkte_spieler2 == 101:
            print("Spieler 2 hat gewonnen! ðŸŽ‰")
            break
        # Spielerwechsel
        aktueller_spieler = 2 if aktueller_spieler == 1 else 1

# Spiel starten
print("ðŸš€ Willkommen zu WÃ¼rfel-101 mit Risiko!")
spiel_101_mit_risiko()
```
