```python
import pygame
import random
import time
import math
from qiskit.circuit.library import HGate
from qiskit import QuantumCircuit
from qiskit.transpiler import generate_preset_pass_manager
from qiskit_aer import AerSimulator
from pygame.math import Vector3
from pygame.math import Vector2

# Pygame initialisieren
pygame.init()
WIDTH, HEIGHT = 1000, 700
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("üé≤ W√ºrfel-101 mit 3D-Animation")

# Farben
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
BLUE = (30, 60, 120)
RED = (200, 50, 50)
GREEN = (50, 180, 50)
GOLD = (255, 215, 0)

# Schriftarten
font_large = pygame.font.SysFont("Arial", 48, bold=True)
font_medium = pygame.font.SysFont("Arial", 32)
font_small = pygame.font.SysFont("Arial", 24)

# 3D-W√ºrfel-Klasse
class Dice3D:
    def __init__(self):
        self.vertices = [
            Vector3(-1, -1, -1),  # 0
            Vector3(1, -1, -1),   # 1
            Vector3(1, 1, -1),    # 2
            Vector3(-1, 1, -1),   # 3
            Vector3(-1, -1, 1),   # 4
            Vector3(1, -1, 1),    # 5
            Vector3(1, 1, 1),     # 6
            Vector3(-1, 1, 1)     # 7
        ]
        self.faces = [
            (0, 1, 2, 3),  # Back
            (4, 5, 6, 7),   # Front
            (0, 1, 5, 4),   # Bottom
            (2, 3, 7, 6),   # Top
            (0, 3, 7, 4),   # Left
            (1, 2, 6, 5)    # Right
        ]
        self.dot_positions = {
            1: [(0, 0, 0.8)],
            2: [(-0.5, -0.5, 0.8), (0.5, 0.5, 0.8)],
            3: [(-0.5, -0.5, 0.8), (0, 0, 0.8), (0.5, 0.5, 0.8)],
            4: [(-0.5, -0.5, 0.8), (-0.5, 0.5, 0.8), 
                (0.5, -0.5, 0.8), (0.5, 0.5, 0.8)],
            5: [(-0.5, -0.5, 0.8), (-0.5, 0.5, 0.8), (0, 0, 0.8),
                (0.5, -0.5, 0.8), (0.5, 0.5, 0.8)],
            6: [(-0.5, -0.5, 0.8), (-0.5, 0, 0.8), (-0.5, 0.5, 0.8),
                (0.5, -0.5, 0.8), (0.5, 0, 0.8), (0.5, 0.5, 0.8)]
        }
        self.angle_x = 0
        self.angle_y = 0
        self.angle_z = 0
        self.scale = 80
        self.pos = Vector3(WIDTH//2, HEIGHT//2, 0)
        self.value = 1
        self.rolling = False
        self.roll_start_time = 0
        self.roll_duration = 1.5  # Sekunden

    def rotate(self, x, y, z):
        self.angle_x += x
        self.angle_y += y
        self.angle_z += z

    def project(self, point):
        # 3D -> 2D Projektion
        rotated = point.rotate_x(self.angle_x).rotate_y(self.angle_y).rotate_z(self.angle_z)
        projected = Vector2(rotated.x * self.scale + self.pos.x, 
                           -rotated.y * self.scale + self.pos.y)
        return projected

    def draw(self, surface):
        # W√ºrfel zeichnen
        for face in self.faces:
            points = [self.project(self.vertices[i]) for i in face]
            pygame.draw.polygon(surface, WHITE, points)
            pygame.draw.polygon(surface, BLACK, points, 2)

        # Punkte zeichnen
        for dot in self.dot_positions[self.value]:
            dot_pos = Vector3(*dot)
            rotated_dot = dot_pos.rotate_x(self.angle_x).rotate_y(self.angle_y).rotate_z(self.angle_z)
            projected_dot = self.project(rotated_dot + Vector3(0, 0, 0.8))
            pygame.draw.circle(surface, BLACK, (int(projected_dot.x), int(projected_dot.y)), 10)

    def roll(self, target_value):
        self.rolling = True
        self.roll_start_time = time.time()
        self.target_value = target_value

    def update(self):
        if self.rolling:
            elapsed = time.time() - self.roll_start_time
            if elapsed < self.roll_duration:
                # W√§hrend der Animation schnell rotieren
                self.rotate(0.3, 0.4, 0.2)
                # Tempor√§r zuf√§lligen Wert anzeigen
                self.value = random.randint(1, 6)
            else:
                self.rolling = False
                self.value = self.target_value

# Quanten-W√ºrfel-Funktion
def wuerfeln():
    backend = AerSimulator(method='extended_stabilizer')
    hadamard_gate = HGate()

    circuit = QuantumCircuit(3, 3)
    for i in range(3):
        circuit.append(hadamard_gate, [i])
    circuit.measure([0, 1, 2], [0, 1, 2])

    pm = generate_preset_pass_manager(backend=backend)
    isa_circuit = pm.run(circuit)

    job = backend.run(isa_circuit, shots=1, memory=True)
    result = job.result()
    memory = result.get_memory()[0]
    dezimal = int(memory, 2)

    if dezimal in [0, 7]:
        return wuerfeln()  # Rekursion bei ung√ºltigem Wurf
    else:
        return dezimal

# Hauptspiel-Logik
def main():
    punkte_spieler1 = 0
    punkte_spieler2 = 0
    runden_punkte = 0
    aktueller_spieler = 1
    
    dice = Dice3D()
    current_value = 1
    rolling = False
    show_result = False
    result_message = ""
    result_timer = 0
    
    clock = pygame.time.Clock()
    running = True
    
    while running:
        dt = clock.tick(60) / 1000.0  # Delta-Zeit in Sekunden
        
        # Events verarbeiten
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            
            if event.type == pygame.MOUSEBUTTONDOWN and not rolling and not show_result:
                mouse_pos = pygame.mouse.get_pos()
                
                # W√ºrfeln-Button
                if WIDTH//2 - 150 <= mouse_pos[0] <= WIDTH//2 - 30 and 550 <= mouse_pos[1] <= 600:
                    target_value = wuerfeln()
                    dice.roll(target_value)
                    rolling = True
                
                # Sichern-Button
                elif WIDTH//2 + 30 <= mouse_pos[0] <= WIDTH//2 + 150 and 550 <= mouse_pos[1] <= 600:
                    if aktueller_spieler == 1:
                        punkte_spieler1 += runden_punkte
                    else:
                        punkte_spieler2 += runden_punkte
                    
                    runden_punkte = 0
                    aktueller_spieler = 2 if aktueller_spieler == 1 else 1
        
        # Spiel-Logik aktualisieren
        if rolling:
            dice.update()
            if not dice.rolling:  # Animation beendet
                rolling = False
                current_value = dice.value
                
                if current_value == 1:
                    result_message = f"Spieler {aktueller_spieler} hat eine 1 gew√ºrfelt! Runde verloren!"
                    runden_punkte = 0
                    aktueller_spieler = 2 if aktueller_spieler == 1 else 1
                else:
                    runden_punkte += current_value
                    result_message = f"Spieler {aktueller_spieler} hat eine {current_value} gew√ºrfelt!"
                    if (aktueller_spieler == 1 and punkte_spieler1 + runden_punkte > 101) or \
                       (aktueller_spieler == 2 and punkte_spieler2 + runden_punkte > 101):
                        result_message = "Mehr als 101 Punkte! Runde beendet."
                        runden_punkte = 0
                        aktueller_spieler = 2 if aktueller_spieler == 1 else 1
                
                show_result = True
                result_timer = time.time()
        
        if show_result and time.time() - result_timer > 2:  # Nachricht f√ºr 2 Sekunden anzeigen
            show_result = False
        
        # Gewinnpr√ºfung
        if punkte_spieler1 >= 101 or punkte_spieler2 >= 101:
            winner = 1 if punkte_spieler1 >= 101 else 2
            result_message = f"Spieler {winner} hat gewonnen! üéâ"
            show_result = True
            result_timer = time.time()
            # Nach 3 Sekunden beenden
            if time.time() - result_timer > 3:
                running = False
        
        # Zeichnen
        screen.fill(BLUE)
        
        # Spielstatus anzeigen
        spieler_text = font_large.render(f"Spieler {aktueller_spieler} ist am Zug", True, GOLD)
        screen.blit(spieler_text, (WIDTH//2 - spieler_text.get_width()//2, 30))
        
        punkte_text = font_medium.render(f"Spieler 1: {punkte_spieler1} | Spieler 2: {punkte_spieler2}", True, WHITE)
        screen.blit(punkte_text, (WIDTH//2 - punkte_text.get_width()//2, 100))
        
        runde_text = font_medium.render(f"Aktuelle Runde: {runden_punkte} Punkte", True, WHITE)
        screen.blit(runde_text, (WIDTH//2 - runde_text.get_width()//2, 150))
        
        # 3D-W√ºrfel zeichnen
        dice.draw(screen)
        
        # Buttons
        if not rolling and not show_result:
            pygame.draw.rect(screen, GREEN, (WIDTH//2 - 150, 550, 120, 50), 0, 10)
            pygame.draw.rect(screen, RED, (WIDTH//2 + 30, 550, 120, 50), 0, 10)
            
            wuerfeln_text = font_small.render("W√ºrfeln", True, BLACK)
            sichern_text = font_small.render("Sichern (+)", True, BLACK)
            
            screen.blit(wuerfeln_text, (WIDTH//2 - 150 + 60 - wuerfeln_text.get_width()//2, 
                                       550 + 25 - wuerfeln_text.get_height()//2))
            screen.blit(sichern_text, (WIDTH//2 + 30 + 60 - sichern_text.get_width()//2, 
                                      550 + 25 - sichern_text.get_height()//2))
        
        # Ergebnis-Nachricht
        if show_result:
            msg_surface = font_medium.render(result_message, True, WHITE)
            pygame.draw.rect(screen, (0, 0, 0, 150), (WIDTH//2 - 200, 400, 400, 50))
            screen.blit(msg_surface, (WIDTH//2 - msg_surface.get_width()//2, 415))
        
        pygame.display.flip()

    pygame.quit()

if __name__ == "__main__":
    main()
```

```python
import pygame
import random
import time
import math
from qiskit.circuit.library import HGate
from qiskit import QuantumCircuit
from qiskit.transpiler import generate_preset_pass_manager
from qiskit_aer import AerSimulator
from pygame.locals import *
from OpenGL.GL import *
from OpenGL.GLU import *

# Pygame und OpenGL initialisieren
pygame.init()
WIDTH, HEIGHT = 1000, 700
screen = pygame.display.set_mode((WIDTH, HEIGHT), DOUBLEBUF | OPENGL)
pygame.display.set_caption("üé≤ W√ºrfel-101 mit echtem 3D-W√ºrfel")

# OpenGL Setup
glEnable(GL_DEPTH_TEST)
glEnable(GL_LIGHTING)
glEnable(GL_LIGHT0)
glEnable(GL_COLOR_MATERIAL)
glColorMaterial(GL_FRONT_AND_BACK, GL_AMBIENT_AND_DIFFUSE)

# Lichtquelle konfigurieren
glLightfv(GL_LIGHT0, GL_POSITION, [1, 1, 1, 0])
glLightfv(GL_LIGHT0, GL_DIFFUSE, [1.0, 1.0, 1.0, 1.0])
glLightfv(GL_LIGHT0, GL_AMBIENT, [0.2, 0.2, 0.2, 1.0])

# Projektion einrichten
glMatrixMode(GL_PROJECTION)
glLoadIdentity()
gluPerspective(45, (WIDTH / HEIGHT), 0.1, 50.0)
glMatrixMode(GL_MODELVIEW)
glLoadIdentity()
gluLookAt(0, 0, 5, 0, 0, 0, 0, 1, 0)

# Farben
WHITE = (1, 1, 1, 1)
BLACK = (0, 0, 0, 1)
RED = (0.8, 0.2, 0.2, 1)
GREEN = (0.2, 0.8, 0.2, 1)
BLUE = (0.2, 0.2, 0.8, 1)
GOLD = (1, 0.84, 0, 1)

# Schriftarten
font_large = pygame.font.SysFont("Arial", 48, bold=True)
font_medium = pygame.font.SysFont("Arial", 32)
font_small = pygame.font.SysFont("Arial", 24)

class Dice3D:
    def __init__(self):
        self.vertices = [
            [-1, -1, -1],  # 0
            [1, -1, -1],   # 1
            [1, 1, -1],    # 2
            [-1, 1, -1],   # 3
            [-1, -1, 1],   # 4
            [1, -1, 1],    # 5
            [1, 1, 1],     # 6
            [-1, 1, 1]     # 7
        ]
        self.faces = [
            [0, 1, 2, 3],  # Back
            [4, 5, 6, 7],   # Front
            [0, 1, 5, 4],   # Bottom
            [2, 3, 7, 6],   # Top
            [0, 3, 7, 4],   # Left
            [1, 2, 6, 5]    # Right
        ]
        self.face_colors = [
            [0.9, 0.9, 0.9],  # Back
            [0.9, 0.9, 0.9],   # Front
            [0.9, 0.9, 0.9],   # Bottom
            [0.9, 0.9, 0.9],   # Top
            [0.9, 0.9, 0.9],   # Left
            [0.9, 0.9, 0.9]    # Right
        ]
        self.dot_positions = {
            1: [(0, 0, 1.01)],
            2: [(-0.5, -0.5, 1.01), (0.5, 0.5, 1.01)],
            3: [(-0.5, -0.5, 1.01), (0, 0, 1.01), (0.5, 0.5, 1.01)],
            4: [(-0.5, -0.5, 1.01), (-0.5, 0.5, 1.01), 
                (0.5, -0.5, 1.01), (0.5, 0.5, 1.01)],
            5: [(-0.5, -0.5, 1.01), (-0.5, 0.5, 1.01), (0, 0, 1.01),
                (0.5, -0.5, 1.01), (0.5, 0.5, 1.01)],
            6: [(-0.5, -0.5, 1.01), (-0.5, 0, 1.01), (-0.5, 0.5, 1.01),
                (0.5, -0.5, 1.01), (0.5, 0, 1.01), (0.5, 0.5, 1.01)]
        }
        self.angle_x = 0
        self.angle_y = 0
        self.angle_z = 0
        self.pos = [0, 0, 0]
        self.value = 1
        self.rolling = False
        self.roll_start_time = 0
        self.roll_duration = 2.0
        self.rotation_speed = 0
        self.target_value = 1
        self.bounce_height = 0
        self.bounce_speed = 0
        self.bouncing = False

    def draw(self):
        glPushMatrix()
        glTranslatef(self.pos[0], self.pos[1] - self.bounce_height/10, self.pos[2])
        glRotatef(self.angle_x, 1, 0, 0)
        glRotatef(self.angle_y, 0, 1, 0)
        glRotatef(self.angle_z, 0, 0, 1)
        
        # W√ºrfel zeichnen
        glBegin(GL_QUADS)
        for i, face in enumerate(self.faces):
            glColor3fv(self.face_colors[i])
            for vertex in face:
                glVertex3fv(self.vertices[vertex])
        glEnd()
        
        # Kanten zeichnen
        glColor3f(0, 0, 0)
        glLineWidth(2)
        glBegin(GL_LINES)
        for face in self.faces:
            for i in range(4):
                glVertex3fv(self.vertices[face[i]])
                glVertex3fv(self.vertices[face[(i+1)%4]])
        glEnd()
        
        # Punkte zeichnen (nur wenn nicht in der Roll-Phase oder wenn der W√ºrfel stillsteht)
        if not self.rolling or (self.rolling and time.time() - self.roll_start_time > self.roll_duration * 0.9):
            glColor3f(0, 0, 0)
            for dot in self.dot_positions[self.value]:
                glPushMatrix()
                glTranslatef(dot[0], dot[1], dot[2])
                gluSphere(gluNewQuadric(), 0.15, 16, 16)
                glPopMatrix()
        
        glPopMatrix()

    def roll(self, target_value):
        self.rolling = True
        self.roll_start_time = time.time()
        self.target_value = target_value
        self.rotation_speed = random.uniform(15, 25)
        self.bouncing = True
        self.bounce_height = 0
        self.bounce_speed = 25

    def update(self):
        if self.rolling:
            elapsed = time.time() - self.roll_start_time
            progress = elapsed / self.roll_duration
            
            # Bounce-Effekt
            if self.bouncing:
                self.bounce_height += self.bounce_speed
                self.bounce_speed -= 1.5
                if self.bounce_height <= 0:
                    self.bounce_height = 0
                    if progress > 0.7:
                        self.bouncing = False
                    else:
                        self.bounce_speed = 15 * (1 - progress)
            
            # Rotationsgeschwindigkeit verlangsamen
            current_speed = self.rotation_speed * (1 - progress**2)
            
            # Rotation aktualisieren
            self.angle_x += current_speed * 0.3 * math.sin(progress * 10)
            self.angle_y += current_speed * 0.4 * math.cos(progress * 8)
            self.angle_z += current_speed * 0.2 * math.sin(progress * 12)
            
            # Tempor√§r zuf√§lligen Wert anzeigen
            if progress < 0.7:
                self.value = random.randint(1, 6)
            else:
                if random.random() < 0.3:
                    self.value = self.target_value
            
            if elapsed >= self.roll_duration:
                self.rolling = False
                self.value = self.target_value
                self.bouncing = False
                self.bounce_height = 0
                # W√ºrfel auf richtige Seite ausrichten
                self.align_to_face()

    def align_to_face(self):
        # W√ºrfel so drehen, dass die gew√ºrfelte Zahl nach vorne zeigt
        target_rotations = {
            1: (0, 0, 0),        # Front (Standard)
            2: (90, 0, 0),       # Oben
            3: (0, 90, 0),       # Rechts
            4: (0, -90, 0),      # Links
            5: (-90, 0, 0),      # Unten
            6: (180, 0, 0)       # Hinten
        }
        self.angle_x, self.angle_y, self.angle_z = target_rotations[self.value]

def wuerfeln():
    backend = AerSimulator(method='extended_stabilizer')
    hadamard_gate = HGate()

    circuit = QuantumCircuit(3, 3)
    for i in range(3):
        circuit.append(hadamard_gate, [i])
    circuit.measure([0, 1, 2], [0, 1, 2])

    pm = generate_preset_pass_manager(backend=backend)
    isa_circuit = pm.run(circuit)

    job = backend.run(isa_circuit, shots=1, memory=True)
    result = job.result()
    memory = result.get_memory()[0]
    dezimal = int(memory, 2)

    if dezimal in [0, 7]:
        return wuerfeln()
    else:
        return dezimal

def draw_text(text, font, color, x, y):
    text_surface = font.render(text, True, color)
    text_data = pygame.image.tostring(text_surface, "RGBA", True)
    glWindowPos2d(x, HEIGHT - y - font.get_height())
    glDrawPixels(text_surface.get_width(), text_surface.get_height(), 
                GL_RGBA, GL_UNSIGNED_BYTE, text_data)

def main():
    punkte_spieler1 = 0
    punkte_spieler2 = 0
    runden_punkte = 0
    aktueller_spieler = 1
    
    dice = Dice3D()
    current_value = 1
    rolling = False
    show_result = False
    result_message = ""
    result_timer = 0
    
    clock = pygame.time.Clock()
    running = True
    
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            
            if event.type == pygame.MOUSEBUTTONDOWN and not rolling and not show_result:
                mouse_pos = pygame.mouse.get_pos()
                
                # W√ºrfeln-Button
                if WIDTH//2 - 150 <= mouse_pos[0] <= WIDTH//2 - 30 and 550 <= mouse_pos[1] <= 600:
                    target_value = wuerfeln()
                    dice.roll(target_value)
                    rolling = True
                
                # Sichern-Button
                elif WIDTH//2 + 30 <= mouse_pos[0] <= WIDTH//2 + 150 and 550 <= mouse_pos[1] <= 600:
                    if aktueller_spieler == 1:
                        punkte_spieler1 += runden_punkte
                    else:
                        punkte_spieler2 += runden_punkte
                    
                    runden_punkte = 0
                    aktueller_spieler = 2 if aktueller_spieler == 1 else 1
        
        # Spiel-Logik aktualisieren
        if rolling:
            dice.update()
            if not dice.rolling:
                rolling = False
                current_value = dice.value
                
                if current_value == 1:
                    result_message = f"Spieler {aktueller_spieler} hat eine 1 gew√ºrfelt! Runde verloren!"
                    runden_punkte = 0
                    aktueller_spieler = 2 if aktueller_spieler == 1 else 1
                else:
                    runden_punkte += current_value
                    result_message = f"Spieler {aktueller_spieler} hat eine {current_value} gew√ºrfelt!"
                    if (aktueller_spieler == 1 and punkte_spieler1 + runden_punkte > 101) or \
                       (aktueller_spieler == 2 and punkte_spieler2 + runden_punkte > 101):
                        result_message = "Mehr als 101 Punkte! Runde beendet."
                        runden_punkte = 0
                        aktueller_spieler = 2 if aktueller_spieler == 1 else 1
                
                show_result = True
                result_timer = time.time()
        
        if show_result and time.time() - result_timer > 2:
            show_result = False
        
        # Gewinnpr√ºfung
        if punkte_spieler1 >= 101 or punkte_spieler2 >= 101:
            winner = 1 if punkte_spieler1 >= 101 else 2
            result_message = f"Spieler {winner} hat gewonnen! üéâ"
            show_result = True
            result_timer = time.time()
            if time.time() - result_timer > 3:
                running = False
        
        # Zeichnen
        glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
        glLoadIdentity()
        gluLookAt(0, 0, 5, 0, 0, 0, 0, 1, 0)
        
        # 3D-W√ºrfel zeichnen
        dice.draw()
        
        # 2D-UI Elemente
        glMatrixMode(GL_PROJECTION)
        glPushMatrix()
        glLoadIdentity()
        gluOrtho2D(0, WIDTH, HEIGHT, 0)
        glMatrixMode(GL_MODELVIEW)
        glPushMatrix()
        glLoadIdentity()
        
        # Spielstatus anzeigen
        draw_text(f"Spieler {aktueller_spieler} ist am Zug", font_large, (255, 215, 0, 255), WIDTH//2 - font_large.size(f"Spieler {aktueller_spieler} ist am Zug")[0]//2, 30)
        draw_text(f"Spieler 1: {punkte_spieler1} | Spieler 2: {punkte_spieler2}", font_medium, (255, 255, 255, 255), WIDTH//2 - font_medium.size(f"Spieler 1: {punkte_spieler1} | Spieler 2: {punkte_spieler2}")[0]//2, 100)
        draw_text(f"Aktuelle Runde: {runden_punkte} Punkte", font_medium, (255, 255, 255, 255), WIDTH//2 - font_medium.size(f"Aktuelle Runde: {runden_punkte} Punkte")[0]//2, 150)
        
        # Buttons
        if not rolling and not show_result:
            # W√ºrfeln-Button
            glColor4f(0.2, 0.8, 0.2, 1.0)
            glBegin(GL_QUADS)
            glVertex2f(WIDTH//2 - 150, 550)
            glVertex2f(WIDTH//2 - 30, 550)
            glVertex2f(WIDTH//2 - 30, 600)
            glVertex2f(WIDTH//2 - 150, 600)
            glEnd()
            
            # Sichern-Button
            glColor4f(0.8, 0.2, 0.2, 1.0)
            glBegin(GL_QUADS)
            glVertex2f(WIDTH//2 + 30, 550)
            glVertex2f(WIDTH//2 + 150, 550)
            glVertex2f(WIDTH//2 + 150, 600)
            glVertex2f(WIDTH//2 + 30, 600)
            glEnd()
            
            # Button Text
            draw_text("W√ºrfeln", font_small, (0, 0, 0, 255), WIDTH//2 - 120 - font_small.size("W√ºrfeln")[0]//2, 565)
            draw_text("Sichern (+)", font_small, (0, 0, 0, 255), WIDTH//2 + 60 - font_small.size("Sichern (+)")[0]//2, 565)
        
        # Ergebnis-Nachricht
        if show_result:
            text_width = font_medium.size(result_message)[0]
            glColor4f(0, 0, 0, 0.6)
            glBegin(GL_QUADS)
            glVertex2f(WIDTH//2 - text_width//2 - 20, 400)
            glVertex2f(WIDTH//2 + text_width//2 + 20, 400)
            glVertex2f(WIDTH//2 + text_width//2 + 20, 450)
            glVertex2f(WIDTH//2 - text_width//2 - 20, 450)
            glEnd()
            draw_text(result_message, font_medium, (255, 255, 255, 255), WIDTH//2 - text_width//2, 415)
        
        glMatrixMode(GL_PROJECTION)
        glPopMatrix()
        glMatrixMode(GL_MODELVIEW)
        glPopMatrix()
        
        pygame.display.flip()
        clock.tick(60)

    pygame.quit()

if __name__ == "__main__":
    main()
```
