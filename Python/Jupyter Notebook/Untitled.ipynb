```python
import random

# 10.000 Zufallswerte (0 oder 1)
results = [random.randint(0, 1) for _ in range(10000)]


```

```python
from statsmodels.sandbox.stats.runs import runstest_1samp

# Der Runs-Test prüft, ob die Reihenfolge der Werte zufällig ist
z_stat, p_value = runstest_1samp(results, correction=False)
print(f"Z-Statistik: {z_stat:.3f}, p-Wert: {p_value:.3f}")

if p_value < 0.05:
    print("Die Folge ist NICHT zufällig (Runs-Test).")
else:
    print("Die Folge ist mit hoher Wahrscheinlichkeit zufällig (Runs-Test).")
```

```python
import matplotlib.pyplot as plt

plt.bar(['0', '1'], [results.count(0), results.count(1)])
plt.title('Häufigkeit von 0 und 1')
plt.ylim(0, 6000)
plt.xlabel('Wert')
plt.ylabel('Anzahl')
plt.show()
```
