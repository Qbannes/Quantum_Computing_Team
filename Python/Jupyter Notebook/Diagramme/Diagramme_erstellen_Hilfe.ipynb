```python
# Erzeugen von Balkendiagrammen (Bundestagswahl 2021)
x = ["CDU", "Grüne", "AfD", "FDP", "Die Linke", "SPD", "CSU", "SSW"]
y = [18.9, 25.7, 10.3, 11.5, 4.9, 14.8, 5.2, 0.1]

# Für die Farbdefinitionen kann man hier nachschlagen:
# https://matplotlib.org/stable/gallery/color/named_colors.html
farben = ["darkblue", "green", "lightblue", "yellow", "darkviolet", "red", "blue", "grey"]

# Achsenbeschreibungen hinzufügen
plt.xlabel("Parteien")
plt.ylabel("Zweitstimmen")

# Balkendiagramm erzeugen
plt.bar(x,y, color=farben)
```

```python
# Es lassen sich auch Kreis- bzw. Tortendiagramme zeichnen
x = [3,7,40,20,3,0.9]

plt.pie(x)
```

```python
# Wenn man die Ausgabe nur auf das Diagramm beschränken möchte:
plt.pie(x)
plt.show()
```

```python
# Diagramme speichern als png
plt.pie(x)
plt.savefig('diagramm.png', format='png')
```

```python
# Kreisdiagramm formatieren
# Erzeugen von Balkendiagrammen (Bundestagswahl 2021)
legende = ["CDU", "SPD", "AfD", "FDP", "Die Linke", "Grüne", "CSU", "SSW"]
x = [18.9, 25.7, 10.3, 11.5, 4.9, 14.8, 5.2, 0.1]
farben = ["darkblue", "red", "lightblue", "yellow", "darkviolet", "green", "blue", "grey"]

plt.pie(x, labels=legende, colors=farben)
plt.show()
```
